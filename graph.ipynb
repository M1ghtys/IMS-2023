```python
import matplotlib.pyplot as plt


input_file_path = 'outputSimulation.txt'

# Read model data
with open(input_file_path, 'r') as file:
    input_text = file.read()


data = {}

# Parse the input text
lines = input_text.strip().split('\n')
for i in range(0, len(lines), 10):
    set_data = {}
    for line in lines[i:i+10]:
        parts = line.split(':')
        path = parts[0].split()[-1]
        delay = float(parts[1])
        set_data[path] = delay
    data[i] = set_data

# Create a line graph for each set of dataset
iterator = 0
for set_num, set_data in data.items():
    char = chr(ord('A')+iterator)
    plt.plot(set_data.keys(), set_data.values(), marker='o', label=f'Set {char}')
    iterator+=1

# Set labels and title
plt.xlabel('Paths')
plt.ylabel('Average Iteration Delay')
plt.title('Average Iteration Delay for Different Paths')


plt.ylim(0, 3)

# Show legend
plt.legend()


plt.show()
```
